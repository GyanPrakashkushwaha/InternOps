### Going with the second columns idea.

```python
df['timeline-1'] = "jan"
df['relation'] = 'something'
df['timeline-2'] = None
```

```python
df
```

```python
df['timeline'][15]
```

```python
def timeline_decode(sample):
    text = sample.split(" ")
    return text[0]
df['timeline-1'] = df['timeline'].apply(timeline_decode)
# df['timeline'].apply(timeline_decode)[15]
```

```python
def timeline_rel(sample):
    text = sample.split(" ")
    if len(text) == 1:
        return "only"
    return text[1]
df['relation'] = df['timeline'].apply(timeline_rel)
```

```python
def timeline2(sample):
    text = sample.split(" ")
    if len(text) == 1:
        return None
    return text[-1]
df['timeline-2'] = df['timeline'].apply(timeline2)
```
